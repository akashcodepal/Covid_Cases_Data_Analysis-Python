# Group Them to see data more clearly

```python
# we want to see all confirmed cases,confirmed... regardless of Date
```

```python
df2=df.groupby(['Country'])[['Country', 'Confirmed' , 'Deaths','Recovered']].sum().reset_index()    
# Group by Country
#then Data we want to see-->country, confirmed, deaths, recovered
# Sum to calculate their sum( by country)
# reset-index is just Giving those 0,1,2,3,4,5 index shown below
```

```python
df2.head(10)
```

```python
# Lets groupby --> First by Country then Group By Date 
df3= df.groupby(['Country','Date'])[['Country','Date','Confirmed','Deaths','Recovered']].sum().reset_index() 
```

```python
df3.head(15)
```

## Masking

```python
# cases >1000
df4 = df3[df3['Confirmed']>1000]
```

```python
df4.head(10)
```

# Matplotlib

## Plot Graph for Each Country 

```python
countries = df4['Country'].unique()
len(countries)
```

```python
# to loop through each country
for i in range(len(countries)):
    c=df3[df3['Country']==countries[i]].reset_index()
    plt.plot(np.arange(0,len(c)),c['Confirmed'],color='b',label='confirmed')
    plt.plot(np.arange(0,len(c)),c['Deaths'],color='g',label='Deaths')
    plt.plot(np.arange(0,len(c)),c['Recovered'],color='r',label='Recovered')
    plt.title(countries[i])
    plt.xlabel('Days since the First suspect')
    plt.ylabel('Number of Cases')
    plt.legend() 
    # This command will force the label to be shown on screen 
    # given while Plotting by plot or scatter or any other..  
    plt.show()
    # show will plot for all ..otherwise all will be displayed on 1 Graph
```
